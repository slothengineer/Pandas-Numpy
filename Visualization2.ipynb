### 1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a powerful data visualization library in Python that is built on top of Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics. Here are five commonly used plots in Seaborn, along with their uses:

1. **Scatter Plot:**
   - **Use:** Scatter plots are used to visualize the relationship between two continuous variables. Each point on the plot represents an observation, making it easy to identify patterns, trends, or clusters.

   ```python
   import seaborn as sns
   import matplotlib.pyplot as plt
   
   sns.scatterplot(x='variable1', y='variable2', data=df)
   plt.show()
   ```

2. **Line Plot:**
   - **Use:** Line plots are useful for visualizing trends in data over a continuous interval or time. They are often used to show how a variable changes over time.

   ```python
   import seaborn as sns
   import matplotlib.pyplot as plt
   
   sns.lineplot(x='time', y='value', data=df)
   plt.show()
   ```

3. **Histogram:**
   - **Use:** Histograms are used to visualize the distribution of a single continuous variable. They provide insights into the underlying data distribution, including the central tendency and spread.

   ```python
   import seaborn as sns
   import matplotlib.pyplot as plt
   
   sns.histplot(data=df, x='variable', bins=30, kde=True)
   plt.show()
   ```

4. **Box Plot (or Box-and-Whisker Plot):**
   - **Use:** Box plots are helpful for visualizing the distribution of a continuous variable across different categories or groups. They display the median, quartiles, and potential outliers in the data.

   ```python
   import seaborn as sns
   import matplotlib.pyplot as plt
   
   sns.boxplot(x='category', y='value', data=df)
   plt.show()
   ```

5. **Heatmap:**
   - **Use:** Heatmaps are used to visualize the correlation matrix between multiple variables. They are especially useful in identifying patterns and relationships in large datasets.

   ```python
   import seaborn as sns
   import matplotlib.pyplot as plt
   
   correlation_matrix = df.corr()
   sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
   plt.show()
   ```

These examples demonstrate just a few of the many possibilities with Seaborn. The library offers a wide range of functions for creating various statistical plots to explore and understand different aspects of your data.

### 2.  Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri_data = sns.load_dataset("fmri")

# Create a line plot using Seaborn
sns.set(style="darkgrid") 

# Plot the line plot
plt.figure(figsize=(10, 6)) 
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)

# Add title and labels
plt.title('FMRI Signal Over Time for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')

# Show the plot
plt.show()
````

### 3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic_data = sns.load_dataset("titanic")

# Create subplots with two box plots
plt.figure(figsize=(12, 6))

# Box plot for age
plt.subplot(1, 2, 1)
sns.boxplot(x="pclass", y="age", data=titanic_data)
plt.title('Box Plot of Age by Pclass')

# Box plot for fare
plt.subplot(1, 2, 2)
sns.boxplot(x="pclass", y="fare", data=titanic_data)
plt.title('Box Plot of Fare by Pclass')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
````

### 4. Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset. 

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot a histogram with hue for the 'cut' column
plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x='price', hue='cut', multiple="stack", kde=True)

# Add title and labels
plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Show the plot
plt.show()
````

### 5. Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris_data = sns.load_dataset("iris")

# Plot a pair plot with hue for the 'species' column
sns.set(style="ticks")  # Set the style of the plot, optional
sns.pairplot(iris_data, hue="species", markers=["o", "s", "D"])

# Add title
plt.suptitle("Pair Plot of Iris Dataset with Hue for Species", y=1.02)

# Show the plot
plt.show()
````

### 6. Use the "flights" dataset from seaborn to plot a heatmap.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for a heatmap
flights_matrix = flights_data.pivot("month", "year", "passengers")

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create a heatmap
sns.heatmap(flights_matrix, annot=True, fmt="d", cmap="YlGnBu")

# Add title and labels
plt.title('Passenger Counts by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
````
